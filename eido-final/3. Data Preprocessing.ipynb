Change the column name (remove space)

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

Check for null value

```python
print('amount of null values in the dataset:')

print(df.isnull().sum())

print('total amount of null values in the dataset: ' + str(df.isnull().sum().sum()))
```

Remove duplicate data

```python
#Drop if the rows are the same and keep first one
df = df.drop_duplicates()
#Drop if there is any duplicate id and keep first one
df = df.drop_duplicates('host_id', keep="first")
df.info()
# duplicate = df['host_id'].duplicated().sum()
# print(duplicate)
```

Dealing with null data

REMOVAL

```python
#Drop specific columns
columns_to_drop = ['id','host_id','host_name','host_identity_verified','lat','long','license','name','neighbourhood','country_code','last_review','calculated_host_listings_count','house_rules','reviews_per_month']
df.drop(columns_to_drop, axis = 1, inplace=True)
```

```python
#Drop rows with missing values in the following columns
#df = df.dropna(columns='','')

#Drop all rows with missing values in any columns
df = df.dropna()
```

```python
print(df.isnull().sum())
print(df.info())
print(df.head())
```

Check if there anything with neighbourhood_group

```python
print(df.loc[:,'neighbourhood_group'].unique())
print(df['neighbourhood_group'].value_counts(ascending=True))
```

We can see that there is one wrong value which is brookln, it should be Brooklyn

Replace brookln to Brooklyn

```python
df['neighbourhood_group'] = df['neighbourhood_group'].replace('brookln', 'Brooklyn')
```

```python
print(df.loc[:,'neighbourhood_group'].unique())
print(df['neighbourhood_group'].value_counts(ascending=True))
```

Check country column

```python
print(df.loc[:,'country'].unique())
print(df['country'].value_counts(ascending=True))
```

```python
#there is only one country so we can just drop it
df.drop('country', axis=1, inplace=True)
```

```python
#print(df.loc[:,'country'].unique())
#print(df['country'].value_counts(ascending=True))
```

Check for instant_bookable

```python
#since there is true and false, we can change it to bool
df['instant_bookable'] = df['instant_bookable'].astype('bool')
print(df.loc[:,'instant_bookable'].unique())
print(df['instant_bookable'].value_counts(ascending=True))
```

Check for cancellation_policy

```python
print(df.loc[:,'cancellation_policy'].unique())
print(df['cancellation_policy'].value_counts(ascending=True))
```

Check for room_type

```python
print(df.loc[:,'room_type'].unique())
print(df['room_type'].value_counts(ascending=True))
```

Replace space with '_'

```python
df['room_type'] = df['room_type'].replace({' ': '_', '/': '_or_'}, regex=True)
print(df['room_type'].value_counts(ascending=True))
```

Check for construction_year

```python
print(df.loc[:,'construction_year'].unique())
print(df['construction_year'].value_counts(ascending=True))
```

Change to year to int

```python
df['construction_year'] = df['construction_year'].astype(int)
```

Check for price and service_fee

```python
#price and service_fee dtype in df.info() is object, convert it to float
#we also need to remove the $ sign from the price

df[['price', 'service_fee']] = df[['price', 'service_fee']].replace({'\$': '', ',': ''}, regex=True).astype(float)
#print(df['price'].head())
has_negative_prices = (df['price'] < 0).any()
has_negative_service_fee = (df['service_fee'] < 0).any()
print(has_negative_prices)
print(has_negative_service_fee)
```

combine price and service_fee

```python
#since they have no negative, we can add it both and create new columns called
#total_price
#total_price = price + service_fee
df['total_price'] = df['price'] + df['service_fee']
df = df.drop(['price', 'service_fee'], axis=1)
print(df['total_price'].head())
```

Check negative value

minimum_nights     
number_of_reviews  
review_rate_number   
availability_365

```python
#columns minimum_nights and number_of_reviews should be int
df = df.astype({'minimum_nights': 'int', 'number_of_reviews': 'int','availability_365': 'int'})
columns_to_check_for_negative_values = ['minimum_nights','number_of_reviews','review_rate_number','availability_365']
has_negative_values = df[columns_to_check_for_negative_values].lt(0).any()
print(has_negative_values)
```

```python
df[columns_to_check_for_negative_values] = df[columns_to_check_for_negative_values].abs()
has_negative_values = df[columns_to_check_for_negative_values].lt(0).any()
print(has_negative_values)
```
